```python
import math
import numpy as np
from scipy.special import factorial, comb, perm
import pprint
import itertools as it
```

```python
def exercise_a():
    # different ways of ordering elements in multisets.

    # the elements of set S = {1,2,5}
    S = np.array([1,2,5])
    # permutations of S
    print(f"S = {S}")
    # number of ways of ordering the elements of S
    print(f"Permutations of S:",factorial(S.size))

#exercise_a()

def exercise_b():
    # elements of set S in groups of pairs 

    #the elements of set S = {j, NaN, 0,3}
    S = np.array(["j", np.nan, 0, 3])
    # p(4,2)
    print(f"S = {S}")
    print(f"the elements of S in groups of pairs:",perm(S.size,2))

#exercise_b()

def exercise_c():
    # elments of multiset S any 5 elements

    # the elements of multiset S = {0,7,0,0,7,7}
    permest = [0,7,0,0,7,7]
    r = 5

    print("0,0,7,7,7")
    outlist = list(set((it.permutations(permest, r))))
    num_permutations1 = factorial(5) / ((factorial(3) * factorial(2)))
    print(len(outlist))
    pprint.pprint(outlist)
    print("0,0,0,7,7")
    num_permutations2 = factorial(5) / ((factorial(3) * factorial(2)))
    perm = num_permutations1 + num_permutations2
    print(perm)

#exercise_c()

def exercise_d():
    # any 6 elements of multiset S
    ## permest = [15,24,x,math.inf,factorial(4),factorial(2),math.inf]
    permest = [15,15,24,24,math.inf,math.inf,2]
    r = 6

    print("15,15,24,24,math.inf,factorial(2)")
    outlist = list(set((it.permutations(permest, r))))
    num_permutations1 = factorial(6) / ((factorial(2) * factorial(2) * factorial(1)*factorial(1)))
    print(len(outlist))
    pprint.pprint(outlist)
    print("15.15.24,math.inf,math.inf,factorial(2)")
    num_permutations2 = factorial(6) / ((factorial(2) * factorial(1) * factorial(2)*factorial(1)))
    print("15,24,24,math.inf,math.inf,factorial(2)")
    num_permutations3 = factorial(6) / ((factorial(1) * factorial(2) * factorial(2)*factorial(1)))
    print("15,15,24,24,math.inf,math.inf")
    num_permutations4 = factorial(6) / ((factorial(2) * factorial(2) * factorial(2)*factorial(1)))
    perm = num_permutations1+num_permutations2+num_permutations3+num_permutations4
    print(perm)
exercise_d()
```
