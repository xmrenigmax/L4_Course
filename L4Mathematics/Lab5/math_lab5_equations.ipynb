```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# briefings

```python
def Briefings():
    def func1(x):
        return 1/x
    # vectorize the function (so that it can take a numpy array as input)
    np.vectorize(func1)
    # create an array of x values from 0.1 to 1000 in steps of 0.1
    x = np.arange(0.1,1000,0.1)
    y = func1(x)
    print(y)

    # plot the function
    x = np.linspace(-100,100,1000)
    y = 1/x
    plt.plot(x,y, 'g:')
    plt.xlabel('x'); plt.ylabel('y')
Briefings()
```
