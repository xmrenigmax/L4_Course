# Vectors in the space

Task A

```python
def taska():
    v = np.array([3,3])
    u = np.array([2,2])
    d = u-v
    plt.quiver(0,0,v[0],v[1],
               angles = 'xy',
               scale_units = 'xy',
               scale = 1,
               color = 'b', label = 'v')
    
    plt.quiver(0, 0, u[0], u[1], angles='xy',  scale_units='xy', scale=1, color='green', label='w')

    plt.quiver(v[0], v[1], d[0], d[1], angles='xy', scale_units='xy', scale=1, color='red', label='d = w - v')
    distance = np.linalg.norm(d)
    print(round(distance**2))
    plt.axis('equal')
    plt.legend()
    plt.grid()

taska()

```

```python
import numpy as np
import matplotlib.pyplot as plt


def taskd():
    # white stone's location
    u = np.array([1, 1, 0])

    # black stones' locations
    v = np.array([1, 0, 0])
    w = np.array([0, 1, 0])
    x = np.array([1, 2, 0])
    y = np.array([2, 1, 0])
    z = np.array([1, 0, 1])
    a = np.array([0, 1, 1])
    b = np.array([1, 2, 1])

    # distance between white and black stones
    distance1 = np.linalg.norm(u - v)
    distance2 = np.linalg.norm(u - w)
    distance3 = np.linalg.norm(u - x)
    distance4 = np.linalg.norm(u - y)
    distance5 = np.linalg.norm(u - z)
    distance6 = np.linalg.norm(u - a)
    distance7 = np.linalg.norm(u - b)

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')

    # scatter to jot down all the points of black stones
    
    


    for i in list([v, w, x, y, z, a, b]):
        ax.plot(0, 0, 0, i[0], i[1], i[2], 'bo-')
        ax.plot(i[0], i[1], i[2], u[0] - i[0], u[1] - i[1], u[2] - i[2], 'ro-')
        # loops through all distances to print
    for i in [distance1, distance2, distance3, distance4, distance5, distance6, distance7]:
        print(i**2)

    ax.legend(['v', 'w', 'x', 'y', 'z', 'a', 'b'])
    ax.axis('equal')
    ax.grid()

taskd()
```
