```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

global a
a = 5
```

# Task 1

```python
def hidden_card():
    # complex number indicated a position in room
    # real number indicates horizontal position
    # imaginary number indicates vertical position

    # complex number z determines the initial position in room
    # j is the imaginary unit
    # a is the group number
    j = 1j

    z = ((1+j)/math.sqrt(2)*(j**4)*(((1+j)/(abs(1+j)))**a))
    # next position is determined by the complex number u
    # z = z / u 
    # next position is z = z / u again and again
    # moves up to 64 times and the answer is the mean of all the positions
    u = math.sqrt(2)/(1+j)
    pos = []

    for z in range(63):
        z = z / u
        pos.append(z)
        plt.plot(z.real, z.imag, 'ro')
    plt.xlabel('real')
    plt.ylabel('imaginary')
    plt.grid()
    plt.show()
    mean_pos = np.mean(pos)
    exact_pos =np.mean(pos)

    # print the mean position of the 64th position as an integer
    # gain access to position in list, turn into integer coordinates
    mean_pos = int(mean_pos.real), int(mean_pos.imag)
    print("This is the mean position of room 1: {}".format(mean_pos))
    return exact_pos
hidden_card()
```

# Task 2

```python
def second_room(exact_pos):
    # square but double the size of room 1
    # z determines the initial position in room 2 now
    # j is the imaginary unit
    j = 1j
    z =((math.sqrt(2))*(j**4))/(((1+j)/abs(1+j))**a)
    # next position is determined by the complex number u
    u = math.sqrt(2)/(1+j)

    pos = []

    # calculate minimum number of moves to reach the mean position of room 1
    # gets the mean position of room 1
    # tracks minimum number of moves to reach mean position of room 2
    min_moves = 0
    # tracks the current position in room 2
    # moves up to 64 times
    for _ in range(64):
        if np.array_equal((int(z.real), int(z.imag)), exact_pos):
            break
        z = z / u
        pos.append(z)
        min_moves += 1

    print("Minimum number of moves to reach the mean position of room 1 in room 2: {}".format(min_moves))
    return min_moves

exact_pos = hidden_card()
second_room(exact_pos)
```
