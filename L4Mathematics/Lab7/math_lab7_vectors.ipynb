# import




```python
# Install the matplotlib package
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def vector_a():
    w = np.array([1,1])
    alpha_values = [3,2,1,0.5,0]
    for i, value in enumerate(alpha_values):
        alpha = np.sqrt(value)
        v = w*alpha
        plt.plot([0,v[0]], [0,v[1]], label =f"Alpha{i+1}:{alpha:.2f}")
        plt.arrow(0,0,v[0],v[1],head_width=0.05, head_length=0.05)
        print("vector",v)
        print("length for alpha", alpha, "is: ", np.linalg.norm(v))
    plt.legend(fontsize=10, loc="upper right")
    plt.show()

vector_a()
```

# Vector work

```python
def vector_b():
    v0 = np.array([3, 4])
    alpha_values = []
    for i in range(1, 6):
        alpha = np.random.rand()
        alpha_values = np.append(alpha_values, alpha)
        print(alpha)
        for i, value in enumerate(alpha_values):
            w = v0*value
            plt.plot([0,w[0]], [0,w[1]], label =f"Alpha{i+1}:{alpha:.2f}")
            plt.arrow(0,0,w[0],w[1],head_width=0.05, head_length=0.05)
            print("vector",w)
            print("length for alpha", alpha, "is: ", np.linalg.norm(w))
    print(alpha_values)
    plt.legend(fontsize=10, loc="upper right")
    plt.show()

vector_b()
```

# vector (C)

```python
def vector_c():
    w = np.array([-1,-1,-2])
    alpha_values = [1,1/math.sqrt(2),1/math.sqrt(4),1/math.sqrt(6),0]
    for i, value in enumerate(alpha_values):
        alpha = np.sqrt(value)
        v = w*alpha
        plt.plot([0,v[0]], [0,v[1]], [0,v[2]], label =f"Alpha{i+1}:{alpha:.2f}")
        plt.arrow(0,0,w[0],w[1],head_width=0.05, head_length=0.05)
        print("length for alpha", alpha, "is: ", np.linalg.norm(v))
    plt.legend(fontsize=10, loc="upper right")
    plt.show()

vector_c()
        
    
```
